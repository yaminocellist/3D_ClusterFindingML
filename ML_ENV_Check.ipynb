```python
# Machine Environment Checking
# For Apple Silicon
import os
os.environ["PYTORCH_ENABLE_MPS_FALLBACK"] = "1"

import torch
# Tells MPS to use a more conservative memory allocation strategy
# os.environ["PYTORCH_MPS_HIGH_WATERMARK_RATIO"] = "0.0"

if torch.backends.mps.is_available():
    print("MPS is available! Your M1 GPU is ready for training.")
    device = torch.device("mps")
else:
    print("MPS not available. Check your PyTorch installation.")

import torch.nn as nn
import torch.optim as optim

# Define a simple toy network
class MyNeuralNetwork(nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = nn.Linear(3, 1)

    def forward(self, x):
        return self.linear(x)

# Now this line will work perfectly:
model = MyNeuralNetwork().to(device)

# 1. Move a tensor to MPS
x = torch.randn(3, 3)
x_mps = x.to(device)

# 2. Move a model to MPS
model = MyNeuralNetwork().to(device)

y_true = torch.randn(3, 1).to(device)

criterion = nn.MSELoss()    # MSE is Mean Square Error
optimizer = optim.SGD(model.parameters(), lr=0.01)

print("\n--- Starting Model Assessment ---")
for epoch in range(5):
    optimizer.zero_grad()
    y_pred = model(x_mps)
    loss = criterion(y_pred, y_true)
    loss.backward()
    optimizer.step()
    print(f"Epoch [{epoch+1}/5] | Loss: {loss.item():.4f} | Device: {y_pred.device}")

print("\nAssessment Complete! If the loss decreased and the device says 'mps:0', your GPU is perfectly configured for 3D Cluster ML training.")
```
